```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
raw = pd.read_csv ("../datasets/eloDadosMarço.csv", header=0)

## Filtering variables
raw = raw[['momento','ta_real','ampa_real', 'pa', 'p3', 'fi_a', 'freq',]]
raw['momento'] = pd.to_datetime (raw['momento'])
raw = raw.set_index(raw.momento)
raw
```

```python
resampled = raw.resample('B').agg({
    'ta_real': ['mean'],
    'ampa_real': ['mean'],
    'fi_a': ['mean', 'std'],
    'pa': ['mean', 'std'],
    'p3': ['mean']
})
resampled.columns = resampled.columns.map('_'.join)
resampled = resampled.dropna()
```

```python
resampled['balance'] = resampled['pa_mean']/resampled['p3_mean']
resampled
```

```python
#fig, ax1 = plt
plt.scatter(resampled['pa_mean'], resampled['fi_a_std'], s=resampled['pa_std'],alpha=0.5)
plt.xlabel('Potência ativa média')
plt.ylabel('Standard Deviation de Phi')
plt.title ('Março de 2019')
```
