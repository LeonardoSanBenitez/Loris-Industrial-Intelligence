# Electrical database exploration
* Convert index to timestamp
* Resample by hour
* Calculates load factor

```python
raw2 = pd.read_csv ("../datasets/2018 ElectricData5min.csv", header=0)
raw2.head()
```

```python
raw2.describe()
```

```python
raw2.plot.kde(subplots=True, sharex=False, figsize=(5,6), layout=(3,1))
plt.savefig('kde2.svg', format='svg')
```

```python
raw2 = raw2[(raw2['p3']>10)] # outlier automatic remotion
```

```python
raw2['timestamp'] = pd.to_datetime (raw2['timestamp']) # already was in brazilian timezone
raw2 = raw2.set_index(raw2.timestamp)
raw2 = raw2.drop ('timestamp', axis=1)
```

```python
# Resample
resampled2 = raw2.resample('H').agg({
    'voltage': ['mean', 'count'],
    'p3': ['mean', 'max'],
    'cosphi': ['mean']
})

resampled2.columns = resampled2.columns.map('_'.join)
resampled2 = resampled2.dropna()
#resampled2.plot(subplots=True)

resampled2
```

```python
filtered2 = resampled2[(resampled2['voltage_count'] == 12)]
filtered2 = filtered2.drop('voltage_count', axis=1)
```

```python
filtered2['load_factor'] = filtered2['p3_mean']/filtered2['p3_max']
```

```python
filtered2 = filtered2.drop('p3_max', axis=1)
```

# Data integration

```python
merged = pd.merge (filtered1, filtered2,how='inner', left_index=True, right_index=True)
```

```python
## Season dummy variables
merged['is_summer'] = merged.index<='2018-03-20 03:00:00' 
merged['is_fall'] = (merged.index>'2018-03-20 00:00:00') & (merged.index<='2018-06-21 00:00:00')
merged['is_winter'] = (merged.index>'2018-06-21 00:00:00') & (merged.index<='2018-09-22 00:00:00')
merged['is_spring'] = (merged.index>'2018-09-22 00:00:00') & (merged.index<='2018-12-21 00:00:00')
merged['is_summer'] = merged['is_summer'] | (merged.index>'2018-12-21 03:00:00')

merged['is_summer'] = merged['is_summer'].apply(lambda x: 1 if x==True else 0)
merged['is_fall'] = merged['is_fall'].apply(lambda x: 1 if x==True else 0)
merged['is_winter'] = merged['is_winter'].apply(lambda x: 1 if x==True else 0)
merged['is_spring'] = merged['is_spring'].apply(lambda x: 1 if x==True else 0)
```

```python
merged.plot(subplots=True)
```

```python
merged.describe()
# sobraram 8067 pontos, de potenciais 8760
```

```python
## Tipo de dia
# type 0 = dia não letivo
# type 1 = dia administrativo
# type 2 = dia letivo completo 
# dias anômalos foram imputados (meio período, reposição de aula, falha no sistema,etc)

merged['dayType'] = 0

# Create an index of just the date portion of your index (this is the slow step)
dfDays = pd.to_datetime(merged.index.date)

# Create a range of business days over that period
dfBdays = pd.bdate_range(start=merged.index[0].date(), end=merged.index[-1].date())

#Filter the series to just those days contained in the business day range.
merged['dayType'] = dfDays.isin(dfBdays)*2 # todos os bussines days são letivos
```

```python
# Janeiro: inteiro não letivo
merged.loc[merged.index.month == 1,'dayType'] = 0

# Fevereiro: consideirei a primeira semana de aula como "administrativo" (é realista, rsrs)
merged.loc[(merged.index.month == 2) & (merged.index.day == 1),'dayType'] = 0
merged.loc[(merged.index.month == 2) & (merged.index.day == 2),'dayType'] = 0
merged.loc[(merged.index.month == 2) & (merged.index.day == 5),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 6),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 7),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 8),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 9),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 12),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 13),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 14),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 15),'dayType'] = 1
merged.loc[(merged.index.month == 2) & (merged.index.day == 16),'dayType'] = 1

# Março: 23 nao teve aula, 30 feriado
merged.loc[(merged.index.month == 3) & (merged.index.day == 23),'dayType'] = 0
merged.loc[(merged.index.month == 3) & (merged.index.day == 30),'dayType'] = 0

# Abril: 30 feriado
merged.loc[(merged.index.month == 4) & (merged.index.day == 30),'dayType'] = 0

# Maio: 1 feriado
# 19 foi reposição
# de 21 a 31 foi greve dos camioneiros 
merged.loc[(merged.index.month == 5) & (merged.index.day == 1),'dayType'] = 0
merged = merged[~((merged.index.month == 5) & (merged.index.day == 19))]
merged = merged[~((merged.index.month == 5) & (merged.index.day >= 21))]

# junho: 1 feriado, 16 e 23 e 30 reposição, 19 zuado
merged.loc[(merged.index.month == 6) & (merged.index.day == 1),'dayType'] = 0
merged = merged[~((merged.index.month == 6) & (merged.index.day == 16))]
merged = merged[~((merged.index.month == 6) & (merged.index.day == 23))]
merged = merged[~((merged.index.month == 6) & (merged.index.day == 30))]

# julho
#7 foi reposição
# ultima e primeira semana de aula considerei adm
# Férias = adm
# dia 18 foi anômalo, algo zuado aconteceu no ifsc
merged = merged[~((merged.index.month == 7) & (merged.index.day == 7))]
merged.loc[(merged.index.month == 7) & (merged.index.day == 2),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 3),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 4),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 5),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 6),'dayType'] = 1

merged.loc[(merged.index.month == 7) & (merged.index.day == 9),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 10),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 11),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 12),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 13),'dayType'] = 1

merged.loc[(merged.index.month == 7) & (merged.index.day == 16),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 17),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 18),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 19),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 20),'dayType'] = 1

merged.loc[(merged.index.month == 7) & (merged.index.day == 23),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 24),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 25),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 26),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 27),'dayType'] = 1

merged.loc[(merged.index.month == 7) & (merged.index.day == 30),'dayType'] = 1
merged.loc[(merged.index.month == 7) & (merged.index.day == 31),'dayType'] = 1

merged = merged[~((merged.index.month == 7) & (merged.index.day == 18))]

# agosto = nada

# setembro: 7 feriado. dia 15 foi anômalo (apesar de ser feriado)
merged.loc[(merged.index.month == 9) & (merged.index.day == 7),'dayType'] = 0
merged = merged[~((merged.index.month == 9) & (merged.index.day == 15))]

# outubro: 12 feraido. 24 anômalo
merged.loc[(merged.index.month == 10) & (merged.index.day == 12),'dayType'] = 0
merged = merged[~((merged.index.month == 10) & (merged.index.day == 24))]

# novembro: 2, 15, 16 feriado
merged.loc[(merged.index.month == 11) & (merged.index.day == 2),'dayType'] = 0
merged.loc[(merged.index.month == 11) & (merged.index.day == 15),'dayType'] = 0
merged.loc[(merged.index.month == 11) & (merged.index.day == 16),'dayType'] = 0

# dezembro: 24 e 25 e 31 feriado
merged.loc[(merged.index.month == 12) & (merged.index.day == 17),'dayType'] = 1
merged.loc[(merged.index.month == 12) & (merged.index.day == 18),'dayType'] = 1
merged.loc[(merged.index.month == 12) & (merged.index.day == 19),'dayType'] = 1
merged.loc[(merged.index.month == 12) & (merged.index.day == 20),'dayType'] = 1
merged.loc[(merged.index.month == 12) & (merged.index.day == 21),'dayType'] = 1

merged.loc[(merged.index.month == 12) & (merged.index.day == 26),'dayType'] = 1
merged.loc[(merged.index.month == 12) & (merged.index.day == 27),'dayType'] = 1
merged.loc[(merged.index.month == 12) & (merged.index.day == 28),'dayType'] = 1

merged.loc[(merged.index.month == 12) & (merged.index.day == 24),'dayType'] = 0
merged.loc[(merged.index.month == 12) & (merged.index.day == 25),'dayType'] = 0
merged.loc[(merged.index.month == 12) & (merged.index.day == 31),'dayType'] = 0
#merged['isBday'] = merged['isBday'].apply(lambda x: 1 if x==True else 0)
```
