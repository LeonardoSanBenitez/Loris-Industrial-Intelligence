```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

result = pd.read_csv ("potencias_2018.csv", header=0);
```

```python
result['data'] = pd.to_datetime (result['data'])
result = result.set_index(result.data)
data = result[['pa']]
```

```python
resampled = data.resample('H').sum()
resampled['paVar'] = data.resample('H').var()

october = resampled['2018-10-10 00:00:00':'2018-10-12 23:00:00']

night = october.between_time('00:00:00', '06:00:00')
night
```
