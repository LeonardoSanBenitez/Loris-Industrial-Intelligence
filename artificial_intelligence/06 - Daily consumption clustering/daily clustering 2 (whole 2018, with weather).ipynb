```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
raw = pd.read_csv ("../datasets/2018 Merging1h.csv", header=0)
raw.head()
```

```python
## Filtering variables
raw['timestamp'] = pd.to_datetime (raw['timestamp'])
raw = raw.set_index(raw.timestamp)
raw = raw.drop('timestamp', axis=1)
raw
```

# Preprocessing
* Normalizo os dados
* calculo a hora do dia (apenas para plotagem futura, não entra no clustering)

```python
normalized=(raw-raw.min())/(raw.max()-raw.min())
normalized
```

```python
normalized['is_summer'] = normalized['is_summer']/4
normalized['is_winter'] = normalized['is_winter']/4
normalized['is_spring'] = normalized['is_spring']/4
normalized['is_fall'] = normalized['is_fall']/4
```

```python
hour = raw.index.hour
```

```python
from sklearn.cluster import KMeans
sse = {}
for k in range(1, 10, 1):
    model = KMeans(n_clusters=k, max_iter=1000).fit(normalized)
    #data["clusters"] = model.labels_
    #print(data["clusters"])
    sse[k] = model.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

# PCA

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(raw.drop('p3_mean', axis=1))
scaled_data = scaler.transform(raw.drop('p3_mean', axis=1))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)

x_pca = pca.transform(scaled_data)
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=raw['p3_mean'],cmap='plasma', alpha=0.1)
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
# Interretando a composição dos componentes
df_comp = pd.DataFrame(pca.components_,columns=raw.drop('p3_mean', axis=1).columns)
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
