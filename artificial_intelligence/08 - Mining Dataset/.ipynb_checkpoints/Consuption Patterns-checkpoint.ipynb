```python


def check_skewness(col):
    sns.distplot(raw[col] , fit=norm);
    fig = plt.figure()
    res = stats.probplot(raw[col], plot=plt)
    # Get the fitted parameters used by the function
    (mu, sigma) = norm.fit(raw[col])
    print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))
    
    
check_skewness('Flotation Column 01 Level')
```

```python
raw = raw.drop('date',axis=1)
```

```python
normalized=(raw-raw.min())/(raw.max()-raw.min())
```

```python

from sklearn.cluster import KMeans
sse = {}
for k in range(1, 5):
    model = KMeans(n_clusters=k, max_iter=1000).fit(normalized)
    #data["clusters"] = model.labels_
    #print(data["clusters"])
    sse[k] = model.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
model = KMeans(n_clusters=2)
model.fit(normalized)

f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True,figsize=(15,6))
ax1.scatter(normalized['% Silica Feed'],normalized['Flotation Column 01 Level'],c=model.labels_,cmap='rainbow', alpha=0.4)
ax2.scatter(normalized['% Silica Feed'],normalized['Flotation Column 04 Level'],c=model.labels_,cmap='rainbow', alpha=0.4)
ax3.scatter(normalized['% Silica Feed'],normalized['% Iron Concentrate'],c=model.labels_,cmap='rainbow', alpha=0.4)
```
